```python
import pandas as pd
import json
```

```python
data = json.load(open("lab_2_coffee.json"))
df = pd.DataFrame(data["results"])
```

```python
df
```

```python
#extract latitude and longitude
df["latitude"] = df ["geocodes"].apply(lambda x: x["main"] ["latitude"])
df["longitude"] = df ["geocodes"].apply(lambda x: x["main"] ["longitude"])
```

```python
df
```

```python
import folium

#create a base map
center_lat = df["latitude"].mean()
center_lng = df["longitude"].mean()
coffee_map = folium.Map(location=[center_lat, center_lng], zoom_start=18)

#add markers
for idx, row in df.iterrows():
    folium.Marker(
        location=[row["latitude"], row ["longitude"]],
        popup=row["name"]
    ).add_to(coffee_map)

coffee_map
```
